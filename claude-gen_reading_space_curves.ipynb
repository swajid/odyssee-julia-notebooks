# Reading Space Curves 🌀 — seeing the shape hiding inside $\vec r(t)$

A vector-valued function $\vec r(t) = \langle x(t),\, y(t),\, z(t)\rangle$ is just a **moving point**:
$t$ is time, and the three slots tell you where the point is at each instant. The skill this notebook
builds: **predict the shape before plotting** — then rotate the plot to confirm your prediction.

Every curve below is interactive: **drag with the mouse to rotate**.

## The three questions to ask any $\vec r(t)$

1. **Where's the circle?** A $\cos$/$\sin$ pair in two slots means circular motion in that plane.
2. **Where's the climb?** A slot that grows steadily (like plain $t$) is the axis the curve travels along.
3. **What multiplies the circle?** Whatever scales the $\cos$/$\sin$ pair is the **radius** — constant,
   growing like $t$, or growing like $t^2$ each give a totally different silhouette.


```julia
using Plots
plotly()                              # interactive backend: drag any plot to rotate
t = range(0, 4π, length=800)          # two revolutions
t1 = range(0, 2π, length=1200)        # one revolution
println("ready")
```

## 1 · The helix: constant radius, steady climb

$\vec r(t) = \langle \cos t,\ \sin t,\ t\rangle$ — the $\cos/\sin$ pair draws a circle of radius 1
in the $xy$-plane while the bare $t$ lifts the point up the $z$-axis at constant speed.
Result: a coil with **evenly spaced loops of identical radius**.

```julia
plot3d(cos.(t), sin.(t), t,
    lw = 3, legend = false, title = "helix: circle in xy, climb along z",
    xlabel = "x", ylabel = "y", zlabel = "z")
```

**Same idea, different axis.** Move the bare $t$ into the *second* slot:
$\vec r(t) = \langle \cos t,\ t,\ \sin t\rangle$. Now the circle lives in the $x$–$z$ plane and the
curve coils along the **y-axis** — the loops stand upright like wheels. The lesson: *the slot
holding the bare $t$ names the axis the curve travels along.*

```julia
plot3d(cos.(t), t, sin.(t),
    lw = 3, legend = false, title = "same helix, climbing along y instead",
    xlabel = "x", ylabel = "y", zlabel = "z")
```

## 2 · Two dials: radius growth and climb speed

Consider the family $\vec r(t) = \langle t^{p}\cos t,\ t^{p}\sin t,\ t^{q}\rangle$.
Dial $p$ controls **how fast the radius grows**; dial $q$ controls **how the height accumulates**.
Three settings, three unmistakable silhouettes:

### $p=1,\ q=1$ — the cone
Radius $=$ height at every instant, so the curve lies exactly on the cone $z=\sqrt{x^2+y^2}$:
a smooth, even taper.

```julia
plot3d(t .* cos.(t), t .* sin.(t), t,
    lw = 3, legend = false, title = "p=1, q=1: radius = height -> lives on a cone",
    xlabel = "x", ylabel = "y", zlabel = "z")
```

### $p=1,\ q=2$ — pancake start, rocket finish
Early on, $t^2$ is tiny, so the first loops happen while the curve has barely lifted —
they sit **squashed near the bottom**. Later, $t^2$ explodes and the curve races upward.

```julia
plot3d(t .* cos.(t), t .* sin.(t), t .^ 2,
    lw = 3, legend = false, title = "p=1, q=2: flat curls low, then a huge climb",
    xlabel = "x", ylabel = "y", zlabel = "z")
```

### $p=2,\ q=1$ — the tail and the flare
Now the *radius* is $t^2$: nearly zero for a while (the curve rises as a thin **vertical tail**),
then it balloons outward into an enormous final sweep. Compare carefully with the cone —
these two get confused most, but the tail is the giveaway.

```julia
plot3d(t .^ 2 .* cos.(t), t .^ 2 .* sin.(t), t,
    lw = 3, legend = false, title = "p=2, q=1: vertical tail, then the radius explodes",
    xlabel = "x", ylabel = "y", zlabel = "z")
```

## 3 · Rings that wiggle: when $z$ oscillates too

Put trig in **all three** slots: $\vec r(t) = \langle R\cos kt,\ R\sin kt,\ A\cos mt\rangle$.
The $xy$-motion is a ring of radius $R$ traced $k$ times, while $z$ bounces $m$ times with
amplitude $A$. Two knobs decide the look:

- **frequency ratio** $m\!:\!k$ — how many wiggles fit per lap
- **amplitude ratio** $A\!:\!R$ — gentle ripple or giant swing

### Many fast, small wiggles → a crown
$R=4.75,\ k=2,\ A=4,\ m=7$: seven modest bounces spread over two laps.

```julia
plot3d(4.75 .* cos.(2t1), 4.75 .* sin.(2t1), 4 .* cos.(7t1),
    lw = 3, legend = false, title = "ring + 7 fast small wiggles: a crown",
    xlabel = "x", ylabel = "y", zlabel = "z")
```

### Few slow, HUGE swings → a potato chip
$R=7,\ k=1,\ A=49,\ m=2$: one lap, two swings — but the amplitude (49) dwarfs the radius (7),
so the "ring" is stretched into a giant saddle. **Amplitude dominance completely changes the
silhouette even though the formula template is identical.**

```julia
plot3d(7 .* cos.(t1), 7 .* sin.(t1), 49 .* cos.(2t1),
    lw = 3, legend = false, title = "ring + 2 giant swings: a potato chip",
    xlabel = "x", ylabel = "y", zlabel = "z")
```

## 4 · The degenerate relatives

### All three slots linear → a straight line
$\vec r(t) = \langle 0.5t,\ 4t,\ 4.75t\rangle$ — no trig anywhere, so no turning at all.
It's the line through the origin with direction vector $\langle 0.5,\ 4,\ 4.75\rangle$
(read straight off the coefficients — this is 2.5's $\vec r_0 + t\vec v$ with $\vec r_0=\vec 0$).

```julia
plot3d(0.5 .* t1, 4 .* t1, 4.75 .* t1,
    lw = 3, legend = false, title = "all slots linear: a straight line",
    xlabel = "x", ylabel = "y", zlabel = "z")
```

### A growing spiral with a faster clock
$\vec r(t) = \langle 5.25t\sin 2t,\ 5.25t\cos 2t,\ 9.25t\rangle$ — section 2's cone pattern
($p=1,\ q=1$) but the $2t$ inside the trig makes it revolve **twice** per period, and the
coefficients stretch it: radius reaches $5.25\cdot 2\pi \approx 33$ while the height reaches
$9.25 \cdot 2\pi \approx 58$.

```julia
plot3d(5.25 .* t1 .* sin.(2t1), 5.25 .* t1 .* cos.(2t1), 9.25 .* t1,
    lw = 3, legend = false, title = "cone pattern, double speed, stretched tall",
    xlabel = "x", ylabel = "y", zlabel = "z")
```

### A shifted, gently tilted circle
$\vec r(t) = \langle 2.75(\cos t + 1),\ 2.75\sin t,\ 2.75\sin\frac{t}{2}\rangle$ — two tricks at once:
the $+1$ **shifts the circle's center** (so $x$ runs $0$ to $5.5$ instead of $\pm 2.75$), and the
half-frequency $\sin\frac{t}{2}$ lifts the loop through **half** an oscillation per lap — a soft tilt,
not a wiggle. Small modifications, totally different personality.

```julia
plot3d(2.75 .* (cos.(t1) .+ 1), 2.75 .* sin.(t1), 2.75 .* sin.(t1 ./ 2),
    lw = 3, legend = false, title = "shifted circle with a slow half-wave lift",
    xlabel = "x", ylabel = "y", zlabel = "z")
```

## 5 · The bounding-box check (verify without plotting)

Before trusting any plot, estimate the curve's **bounding box** by hand: plug the endpoint of the
$t$-interval (and trig maxima $=\pm$ the amplitude) into each slot. For $0\le t\le 2\pi$:

| curve | $x$ reaches | $y$ reaches | $z$ reaches |
|---|---|---|---|
| line $\langle 0.5t, 4t, 4.75t\rangle$ | $0.5\cdot 2\pi \approx 3.1$ | $\approx 25.1$ | $\approx 29.8$ |
| crown ($R=4.75, A=4$) | $\pm 4.75$ | $\pm 4.75$ | $\pm 4$ |
| chip ($R=7, A=49$) | $\pm 7$ | $\pm 7$ | $\pm 49$ |
| tall spiral | $\pm 33$ (radius $5.25\cdot 2\pi$) | $\pm 33$ | $\to 58$ |
| shifted circle | $0$ to $5.5$ | $\pm 2.75$ | $\pm 2.75$ |

If a plot's axis ranges disagree with this arithmetic, the plot isn't the curve you think it is.
This 30-second habit catches almost every mix-up between similar-looking curves.
